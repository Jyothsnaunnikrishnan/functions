```python
#1. Write a python function that takes a list of numbers as input and retuns the
# sum of all even numbers in the list?
def sum_even_numbers(numbers):
    sum_even = 0
    for i in numbers:
        if i % 2 == 0:
            sum_even += i
    return sum_even
```

```python
#3. Implement a python function that takes a list of intigers and retuns a new
# list containing square of each number?
def square_numbers(numbers):
  squares = []
  for i in numbers:
      squares.append(i**2)
  return squares

square_numbers([1,2,3,4,5])
```

```python
#4. Write a python function to check if a number is prime or not from 1to 200?
def is_prime(number):
    if number < 2:
        return False
    for i in range(2, int(number**0.5) + 1):
        if number % i == 0 or number> 200:
            return False
    return True
```

```python
is_prime(13)
```

```python
#6. Write a generator function in python that yield the powers of 2 up to a
#given exponent?
def power_of_two(exponent):
    for i in range(exponent + 1):
        yield 2**i
```

```python
#7. Implement generator function that reads a file line by line and yields line
#as a string?
def read_file_lines(file_path):
    with open(file_path, 'r') as file:
        for line in file:
            yield line.strip()


```

```python
#8. Use a lambda function in python to sort a list of tuples based on the second
# element of each tuple?
data = [('apple', 3), ('banana', 1), ('cherry', 2)]
sorted_data = sorted(data, key=lambda x: x[1])
print(sorted_data)
```

```python
#9. create a python program that uses map()to convert a list of temperature from
#celcius to farenheit?
celsius_temperatures = [0, 10, 20, 30, 40]
fahrenheit_temperatures = list(map(lambda c: (c * 9/5) + 32, celsius_temperatures))
print(fahrenheit_temperatures)
```
